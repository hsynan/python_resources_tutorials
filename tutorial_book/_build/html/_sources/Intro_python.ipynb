## Jupter Notebook Structure
Jupter Notebooks are build out of cells. If you double click on the text, you'll see that these words are written in a cell that can be edited and moved around and deleted. Cells have different types:

* *Markdown* means you're writting in the cell, not coding. 

* *code* means you're coding inside the cell. We can use a combination of markdown cells and coding cells to help keep track of what's happening in the notebook. 

There are two modes for cells: command (which is running the cell) and edit (self explainitory, the cell is not being run but instead is being edited.)
- to enter command mode, click outside the cell. This will stop any typing/coding from continuing in the cell and allow you to run the cell
- to enter edit mode, double click the cell. Alternativly, you can click the cell then hit enter or return.
  
shortcuts:

* `Ctrl` + `Enter` (mac is `Command` + `Enter`) run selected cells
* `Shift` + `Enter` run the current cell, select below
* `Alt` + `Enter` (mac is `Option` + `Enter`) run the current cell, insert below
* `D + D` (press the key twice) to delete the current cell
* if you're in command mode (click outside the cell), press 'a' to add cell above and 'b' to add cell below

# Conditionals
Conditionals are very common in codeing as it allows us to automate alot of processes. They usually look like this:

if (test):

    If test is true, run
    
elif (second test):

    if the first test is false but second test is true, run
    
else:

      if both tests are false, run this part

````python
x = 20
if x>10:
    print('wow this is greater than 20')
elif x == 10:
    print('this number is 10')
else:
    print('sad this number is smaller than 10')
````

# Loops
Another essential python function is loops. Loops cycle through elements of a list or dataset, one at a time, and applies something to the variable. 

````python
a_list = [1, 2, 3]
for i in a_list: #for every i in a (every entry in a_list)
    print(i)
````

````python
b = [] #let's make an empty list 
for i in a_list: #for every i in a (every entry in a_list)
    new_value = i + 10 # add 10 to every entry in a_list
    b.append(new_value) #use append to 'append' the new value to the list (b)
    
    if len(b) == 3:
        print('this is 3 long')
print(b)
````

## Def loops
Another automation is definitions loops or functions. Functions create a resusable code block that basically creates a function that can be run

````python
def addfive(x): #the function name is called addfive, and to run this function we need to tell it what x is
    return x +5 #if you tell the function what x is, it will add 5 to it
````

````python
addfive(3) #wow! a cell ran and automatically followed the function!
````

````python
def multi_variables(a, b, blue):
    '''
    NOTE: when makeing a def loop, always add a definition to the def for code documentation
    '''
    calculation = 1 + a + (b/blue)
    return(calculation)
````

````python
multi_variables(5, 12, 4) #here, a = 5, b = 12, blue = 4
````
