## Set up directory structure
The processing scripts expect a standardized directory structure.
### RESOURCES
The base directory for the processing scripts, utility functions and resources. Within RESOURCES there are two main directories:
* python: for the processing scrips and utility functions
* workflow_resources: for resource files (e.g. metadata, shapefiles, regridding weights)
### DATASETS
The base director for the SOURCE and PRODUCT files

## Clone Repos
Three repos are needed to run the processing pipelines:
* [READ-EDAB-satdata_processing](https://github.com/NEFSC/READ-EDAB-satdata_processing) (clone as satdata_processing into [RESOURCES]/python)
* [READ-EDAB-Utilities-py](https://github.com/NEFSC/READ-EDAB-Utilities-py/tree/main) (clone as utilities into [RESOURCES]/python)
* [READ_EDAB_Resources](https://github.com/NEFSC/READ_EDAB_Resources/tree/satdata#) (clone as workflow_resources into RESOURCES)

### Make directories
Make the RESOURCES and python directories if it does not exist.  

```
mkdir [RESOURCES]
```
(replace [RESOURCES] with the directory path).  

**Example:**   
```
mkdir /Users/kimberly.hyde/Documents/nadata/RESOURCES/
mkdir /Users/kimberly.hyde/Documents/nadata/RESOURCES/python
```

### Change directory
From a terminal window change the directory to the desired "Resources" path.  
```
cd [python]
```  
(replace [python] with the directory path).    

**Example:** 
```
cd /Users/kimberly.hyde/Documents/nadata/RESOURCES/
```

### Clone repos
```
git clone [link to resource branch] workflow_resources
cd python
git clone [link to utilities branch] utilities
git clone [link to satdata_processing branch] satdata_processing
```  

## Set up the Environment

### Change to the satdata_processing directory
```
cd python/satdata_processing
``` 

### Run the setup_env.sh script
The setup_env script will:
1) Create the conda environemnt from satprocessing.yml  
2) Activate the environment  
3) Install required libraries from requirements.txt  
4) Test some of the installs by checking their version  

```
bash setup_env.sh
```  

## Add Resources to environment.py

### Open: 
/[RESOURCES]/python/utilties/bootstrap/environment.py

### Add:
[RESOURCES] path to "resources_root" list in **def bootstrap_environment()** (~ line 35) 

### Confirm path
Use the following to confirm your python paths replacing [path to utilities/bootstrap] with your actual path 
```
import sys
from pprint import pprint  
sys.path.append('[path to utilities/bootstrap]')
from environment import bootstrap_environment  
pprint(bootstrap_environment())
```

## Set up Environment in new modules or functions
After importing necessary python libraries (e.g. datetime, os), import the environment, then the needed utilities
```
from utilities.bootstrap.environment import bootstrap_environment
env = bootstrap_environment(verbose=False)
from utilities import get_dates, get_source_file_dates 
```
