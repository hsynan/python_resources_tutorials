## Calculations
* **calc_phytosizeclass**: to run the phytoplankton size class model
* **calc_daylength**: to calculate the daylength based on day of year and latitude
* **calc_mapped_pixel_area**: to calculate the area of a pixel
* **calc_primprod**: to run the primary productivity model(s)
* **calc_annual_pp**: to calculate annual total primary production
