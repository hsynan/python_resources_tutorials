# Debugging 101 
Coding is a wonderful, important skill, but it is also notoriously frustrating. Its not uncommon find yourself trying to debug an error for hours, only to have it be a simple error in the end. Luckily, python has pretty detailed error messages. Regardless, you will inevitably find yourself in a debugging purgatory. BUT a simple understanding of common python error messages can help you debug more efficiently and quickly. 

## Resources
* [Python troubleshooting](https://studentrobotics.org/docs/troubleshooting/python)
  * This resource describes some of the common error messages
* [A Guide to Debugging Python code (and why you should learn it)](https://medium.com/@yellalena/a-guide-to-debugging-python-code-and-why-you-should-learn-it-ae30d20419b7)
  * This resource defines the goals of debugging and actually gives an example of working through a bug in code. 
* [Understanding the python traceback](https://realpython.com/python-traceback/)
  * This one is REALLY important, it breaks down how to understand a full python traceback (error message)

LLMs can be a great resource for efficient coding, but it can also sometimes cause further confusion if using it purely to debug an issue. It is still really important to understand the steps to debugging and use a LLM to augment. 

## Workflow for fixing errors in code
1. Identify the error
   * First, identify the category of error. This is where knowing what some of the common error types in python mean. Is it a syntax error? Has the variable not been defined? Etc.
3. Isolate and reproduce
   * Create a minimal/simplified example
   * If in a larger workflow, pull out the few lines throwing the error to isolate it 
4. Research and resolve
   * Google/LLM the issue
   * Stack overflow can be pretty helpful for python 
6. Verify
   * Once fixed, add the resovled code back into the larger workflow to ensure it works and the bug has been resolved
  
**Dont be afraid to ask for help!**
