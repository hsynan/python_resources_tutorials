## What is markdown?
Markdown is a lightweight markup language used to display formatted text, images, and equations alongside your live code. It allows you to document your workflow, explain complex steps, and create professional reports that are readable by both technical and non-technical audiences.

## Advantages and disadvantages
* Pros:
  * Easy to comment/explain code (visually appealing)
  * Can create presentations directly from your ipynb files
* Cons:
  * Must be manually rendered
  * Static
  * Clunky (not streamlined to run headlessly) <br>
[This resource](https://erikjandevries.medium.com/when-and-how-jupyter-notebooks-fail-and-what-to-use-instead-a52c27dbaa4c) outline some specific use cases of when to use (and not to use) jupyter notebooks.  <br>
